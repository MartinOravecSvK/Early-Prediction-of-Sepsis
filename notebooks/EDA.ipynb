```python
import pandas as pd
import glob

csv_files = glob.glob('training/**/*.psv', recursive=True)  # For all .psv files 

# List for the dataframes
dfs = []

# Iterate 
for filename in csv_files:
    df = pd.read_csv(filename, delimiter='|')  # pipe as delimiter
    dfs.append(df)

# Concatenate all the dataframes 
combined_df = pd.concat(dfs, ignore_index=True)

print(combined_df.shape)
```
